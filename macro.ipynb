```python
yearly=macro[macro["Frequency"]=="Yearly"]

df['date_disbursement'] = pd.to_datetime(df['date_disbursement'])
yearly['DateTime'] = pd.to_datetime(yearly['DateTime'])

df['Year'] = df['date_disbursement'].dt.year
yearly['Year'] = yearly['DateTime'].dt.year

merged_df = df.merge(yearly, how='left', left_on='Year', right_on='Year')

for category in yearly['Category'].unique():
    if category in merged_df.columns:
        df[category] = merged_df["Value"]

df = df.drop(columns=['Year'])
df = df.reset_index(drop=True)

df
```

```python
quarterly=macro[macro["Frequency"]=="Quarterly"]
print(quarterly["Category"].value_counts())
```

```python
def expand_quarterly_data(df):
    expanded_rows = []
    for _, row in df.iterrows():
        end_of_quarter = row['DateTime']
        start_of_quarter = end_of_quarter - pd.DateOffset(months=2)
        months = [start_of_quarter + pd.DateOffset(months=i) for i in range(3)]
        for month in months:
            expanded_rows.append({'DateTime': month, 'Value': row['Value'], "Category": row["Category"]})
    return pd.DataFrame(expanded_rows)

expanded_macro = expand_quarterly_data(quarterly)
print(expanded_macro.head(5))
expanded_macro['MonthYear'] = expanded_macro['DateTime'].dt.to_period('M')
print(expanded_macro["MonthYear"].head(5))

df['date_disbursement'] = pd.to_datetime(df['date_disbursement'])
df['MonthYear'] = df['date_disbursement'].dt.to_period('M')
print(df["MonthYear"].head(5))

merged_data = pd.merge(df, expanded_macro, left_on='MonthYear', right_on='MonthYear', how='left')
merged_data = merged_data.drop(columns=['DateTime'])

for category in merged_data['Category'].unique():
    if category in merged_data.columns:
        df[category] = merged_data["Value"]

df=df.drop("MonthYear", axis=1)
df = df.reset_index(drop=True)

df.head(20)
```
