```python
#Adding yearly
yearly=macro[macro["Frequency"]=="Yearly"]

df['date_disbursement'] = pd.to_datetime(df['date_disbursement'])
yearly['DateTime'] = pd.to_datetime(yearly['DateTime'])

df['Year'] = df['date_disbursement'].dt.year
yearly['Year'] = yearly['DateTime'].dt.year

merged_df = df.merge(yearly, how='left', left_on='Year', right_on='Year')

for category in yearly['Category'].unique():
    if category in merged_df.columns:
        df[category] = merged_df["Value"]

df = df.drop(columns=['Year'])
df = df.reset_index(drop=True)

df
```

```python
quarterly=macro[macro["Frequency"]=="Quarterly"]

quarterly.head(20)
```

```python
df['date_disbursement'] = pd.to_datetime(df['date_disbursement'])
quarterly['DateTime'] = pd.to_datetime(quarterly['DateTime'])

quarterly_expanded = quarterly.copy()
quarterly_expanded['Start_Date'] = quarterly_expanded['DateTime'] - pd.DateOffset(months=2)
quarterly_expanded['End_Date'] = quarterly_expanded['DateTime']

merged_df = df.merge(quarterly_expanded, how='left', left_on='date_disbursement', right_on='DateTime')

for category in quarterly_expanded['Category'].unique():
    mask = merged_df['Category'] == category
    df.loc[mask, category] = merged_df.loc[mask, 'Value']

df = df.reset_index(drop=True)

df.head(20)
```
